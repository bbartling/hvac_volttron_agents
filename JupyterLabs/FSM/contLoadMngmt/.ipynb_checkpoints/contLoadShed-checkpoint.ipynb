# Continuous Load Management Concept

1. #### Analyze daily load profile and recognize the power level of
    1. Base load (continuous load)
    1. HVAC load (RTU power) and correlation with weather (OA temperature)
    1. Solar PV power generation
1. #### Continuous load management algorithm concept:
    1. Set a peak load limit
    1. Load shift in the morning if OA temperature is predicted high
    1. Collect current solar PV power data and predict the power for the next hour or two
    1. Calculate power to reach limit in the next hour or two
    1. HVAC load shedding during peak through zone temperature reset (change Zone 1 to 4 setpoint depending on the need)
    1. Direct load control (plug load/lighting)


# Sequence

1. #### Set building baseload:
    1. 15 kW
1. #### Set a peak load limit:
    1. 80kW
1. #### Load shift in the morning if OA temperature is predicted high:
    1. 4:00 am – check weather report, highest OA temperature and time of day
    1. 4:00 am – if “Max OA temp” > 90 Deg F, start to pre-cool the building (run load-shifting algorithm – Occupied Setpoint – 3 Deg F for all zones)
1. #### 8:00 am – if building is pre-cooled, change the Occupied Setpoint to normal occupied setpoint for all zones
1. #### Collect current solar PV power data and predict the power for the next hour or two:
    1. 10:00 am – collect solar PV power generation data (kW) – assuming PV power will be the same in the next hour
1. #### Calculate power to reach limit in the next hour or two:
    1. 10:00 am – collect current whole building power (kW) and RTU power (kW)
    1. 10:00 am – calculate power increase limit = peak load limit (80kW) - current whole building power (65kW)
1. #### HVAC load shedding during peak through zone temperature reset (change Zone 1 to 4 setpoint depending on the need):
    1. 10:00 am – 4:00 pm check every 30 minutes - if “power increase limit for RTU” < 1 stage of RTU power (roughly 15kW), or whole building meter power > 65KW, then start load shedding algorithm (Occupied Setpoint + 3 Deg F one zone at a time). Monitor cooling capacity loop output so that it won’t go to the next stage (0% output – 50% - 75 % - 95% with deadbands)
1. #### Direct load control (plug load/lighting):
    1. 10:00 am – 4:00 pm check every 30 minutes – shut down plug loads when power increase limit < 1.5 kW


```python
import os, sys, inspect, io, random
import datetime as dt
from transitions import *
from transitions.extensions import GraphMachine
from IPython.display import Image, display, display_png
```

```python
class MakeChart():
    def show_graph(self, **kwargs):
        stream = io.BytesIO()
        self.get_graph(**kwargs).draw(stream, prog='dot', format='png')
        display(Image(stream.getvalue()))
```

```python
class ContLoadManager(MakeChart):
    
    def __init__(self):
        self.four_am = False
        self.weather_checked = False
        self.bacnet_releases_complete = False
        self.peak_load_window = False
        self.precool_window = False
        self.building_occupied = False
        self.its_going_to_be_hot = False

        
        self.hour = 0
        self.weather_check_hour = 4
        self.building_occ_hour = 8
        self.peak_load_start_hour = 10
        self.peak_load_start_end = 16
        self.pre_cool_start_hour = 4
        self.pre_cool_start_end = 6
        self.hi_out_temp_setpoint = 80 
        
        

    def check_and_set(self):
        #self.hour = dt.datetime.today().hour
        if self.hour == 4 and self.building_occupied == True and self.weather_checked == False:
            self.check_weather()
            self.weather_checked = False

            
    def check_weather(self): print("Checking open weather map!!!!")

     
        
states=['Standby', 'Check_Weather', 
        'Its_Going_To_Be_Hot', 'Its_Going_To_Be_Cool'
        'Pre_Cool_Start', 'Pre_Cool_End',
        'Peak_Load_Time_Window'
        'Reset_Params']



transitions = [

    # Check open weather map API
    {'trigger': 'checkWeather', 'source': 'Standby', 'dest': 'Check_Weather', 
     'conditions':[ # TRUE BOOLEAN
         'four_am'
                  ],
     'unless':[ # FALSE BOOLEAN
         'bacnet_releases_complete',
         'building_occupied',
         'weather_checked',
         'Pre_Cool_Start',
         'Pre_Cool_End'
     ]},

    # Its going to be hot logic
    {'trigger': 'hotWeatherForecast', 'source': 'Check_Weather', 'dest': 'Its_Going_To_Be_Hot', 
     'conditions':[  # TRUE BOOLEAN
         'its_going_to_be_hot',
         'building_occupied'
                  ],
     'unless':[ # FALSE BOOLEAN
         'bacnet_releases_complete',
         'weather_checked',
         'Pre_Cool_Start',
         'Pre_Cool_End'
     ]},

    # Its going to be cool logic back to standby
    {'trigger': 'coolWeatherForecast', 'source': 'Check_Weather', 'dest': 'Standby', 
     'conditions':[  # TRUE BOOLEAN
         'building_occupied'
                  ],
     'unless':[ # FALSE BOOLEAN
         'its_going_to_be_hot',
         'bacnet_releases_complete',
         'weather_checked',
         'Pre_Cool_Start',
         'Pre_Cool_End'
     ]},


    # Its going to be hot logic, go into precool at 8AM
    {'trigger': 'startPrecool', 'source': 'Its_Going_To_Be_Hot', 'dest': 'Its_Going_To_Be_Hot', 
     'conditions':[  # TRUE BOOLEAN
         'precool_window',
         'building_occupied'
                  ],
     'unless':[ # FALSE BOOLEAN
         'bacnet_releases_complete',
         'weather_checked',
         'Pre_Cool_Start',
         'Pre_Cool_End'
     ]},

]



model = ContLoadManager()
machine = GraphMachine(model=model, states=states, transitions=transitions, initial='Standby')
model.show_graph()
```
